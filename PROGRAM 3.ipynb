```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,10))

#1
x = np.array([1,6])
y = np.array([3,5])
plt.subplot(3, 3, 1)
plt.xlabel("x-axis",color='blue')
plt.ylabel("y-axis",color='blue')
plt.plot(x,y,'o')
plt.title("Title1")


#2
y1 = np.array([3, 8, 1, 10, 5])
y2 = np.array([6, 2, 7, 5,10])
plt.subplot(3, 3, 2)
plt.plot(y1,color='orange', linewidth ='2.5',marker = '*')
plt.plot(y2, linestyle = 'dashed',color='blue')


#3
ypoints = np.array([3, 8, 1, 10])
plt.subplot(3, 3, 3)
plt.plot(ypoints,color='blue', marker = 'o', mfc = 'yellow',mec='black')


#4
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(3, 3, 4)
plt.plot(x, y)
plt.grid()
plt.title("Title2")


#5
x = np.random.randint(50, size=(50))
y = np.random.randint(50, size=(50))
plt.subplot(3, 3, 5)
colors = np.random.randint(50, size=(50))
sizes = 5 * np.random.randint(50, size=(50))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')


#6
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.subplot(3, 3, 6)
plt.bar(x, y, width = 0.1,color='r')


#7
y = np.array([35, 25, 20, 20])
myexplode = [0.2, 0, 0, 0]
mycolors=["red","pink","grey","black"]
mylabels = ["A","B","C","D"]
plt.subplot(3, 3, 7)
plt.pie(y,startangle = 90,explode = myexplode,shadow = True,colors=mycolors,labels = mylabels)
plt.legend()

#8

x = np.random.normal(150, 10, 250)
plt.subplot(3, 3, 8)
plt.hist(x,color='green')

#9
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.subplot(3, 3, 9)
plt.scatter(x, y, c=colors)

plt.suptitle("GRAPH")
plt.subplots_adjust(left=0.1,right=0.9,top=0.9,bottom=0.1,wspace=0.3,hspace=0.3)
plt.show()
```
